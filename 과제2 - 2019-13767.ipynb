```python
import math

axis0 = int(input("행의 갯수: "))
axis1 = int(input("열의 갯수: "))
table = np.zeros((axis0,axis1))


max_data = np.array(df1.iloc[:,3])
max_data = math.ceil(max(max_data))

print(f'단지 규모의 최댓값은: {max_data} 입니다.')


def func_count(x, y, v, z):
    index = []
    for i in range(len(df1)):
        # 연령이 x ~ y 범위 에 있다면,
        if df1.iloc[i,2] >= x and  df1.iloc[i,2] <= y:
            # 단지규모가 v ~ z 범위 에 있다면,
            if df1.iloc[i, 3] > v and df1.iloc[i,3] <= z:
                index.append(i)
    return len(index)
    
for i in range (axis0):
    for j in range (axis1):
        print(f'\n{i}행 {j}번째 열입니다.')
        x = int(input("연령이 x ~ y 일때, 연령 시작점(x)을 입력하세요.: "))
        y = int(input("연령이 x ~ y 일때, 연령 끝점(y)을 입력하세요.: "))
        v = int(input("단지규모가 v ~ z 일때, 단지규모 시작점(v)을 입력하세요.: "))
        z = int(input("단지규모가 v ~ z 일때, 단지규모 끝점(z)을 입력하세요.: \n"))
        count = func_count(x,y,v,z)
        table[i][j] = count
        print(table)
print(table)
```

```python
df1['평당가격X연령'] = df1['평당가격'] * df1['연     령']
df1['단지규모X연령'] = df1['단지규모'] * df1['연     령']
df1
```

```python
cov1 = df1.평당가격X연령.mean() - df1.평당가격.mean() * df1.iloc[:, 2].mean()
cov1
```

```python
std_연령 = df1.iloc[:,2].std()
std_평당가격 = df1.iloc[:,4].std()
corr1 = cov1/std_연령/std_평당가격
print(f'correlation is {corr1}')
```

```python
cov2 = df1.단지규모X연령.mean() - df1.단지규모.mean() * df1.iloc[:, 2].mean()
cov2
```
